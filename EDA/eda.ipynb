```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('hrdata.csv')
df
```

```python
l = list(df['Years At Company'])
t1 = []
pv = 0
for i in range(len(l)):
    if i<=10:
        pv="lowest"
        t1.append(pv)
    elif i<=20:
        pv = "medium"
        t1.append(pv)
    elif i<=30:
        pv = "average"
        t1.append(pv)
    else:
        pv = "highest"
        t1.append(pv)
t1
```

```python
df['Year of Experience'] = t1
df
```
